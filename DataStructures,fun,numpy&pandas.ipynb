```python
##Section A: Data Structures & Control Structures

numbers = [10, 20, 30, 40, 50] #list

# append
numbers.append(60)
print(numbers)

# extend
numbers.extend([70, 80])
print(numbers)

# insert
numbers.insert(2, 25)
print(numbers)

# remove
numbers.remove(40)
print(numbers)

# pop
numbers.pop()
print(numbers)

# index
x = numbers.index(30)
print("Index of 30 is", x)

# count
y = numbers.count(20)
print("Count of 20 is", y)

# sort
numbers.sort()
print(numbers)

# reverse
numbers.reverse()
print(numbers)

# clear
numbers.clear()
print(numbers)
```

Section C: Exploring Pandas

```python
import pandas as pd
import numpy as np

# Student details data

data = {
    "Name": ["Ravi", "Sita", "Rahul", "Priya", "Kiran",
             "Asha", "Vijay", "Meena", "Arun", "Ravi"],

    "Age": [22, 24, 21, 25, 23,
            26, 27, 22, 24, 22],

    "City": ["Hyderabad", "Chennai", "Bangalore", "Mumbai", "Delhi",
             "Pune", "Hyderabad", "Chennai", "Delhi", "Hyderabad"],

    "Salary": [25000, 30000, np.nan, 35000, 28000,
               40000, 45000, 30000, 28000, 25000]
}

df = pd.DataFrame(data)

# Checking dataframe information
df.info()

# statistics
print(df.describe())

# Displaying only Name and City columns
print(df[["Name", "City"]])

# Removing City column
df = df.drop("City", axis=1)

# Replacing missing salary value with average salary(mean)
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())

# Removing duplicate records
df = df.drop_duplicates()

# Final dataframe
print(df)
```

```python
num = int(input("Enter a number: "))

count = 0

for i in range(1, num + 1):
    if num % i == 0:
        count = count + 1

if count == 2:
    print("Prime Number")
else:
    print("Not a Prime Number")
```

```python
import random

a = random.randint(1, 10)
b = random.randint(1, 10)

print("First Num=", a)
print("Second Num =", b)

user_answer = int(input("Enter Product: "))

correct_answer = a * b

if user_answer == correct_answer:
    print("Correct")
else:
    print("Wrong")
    print("Correct Product is", correct_answer)
```

```python
days = 4
tests = 3
medicine = 2400

room_charge = days * 2000
doctor_fee = 1500
lab_charge = tests * 300

total_bill = room_charge + doctor_fee + lab_charge + medicine

if total_bill > 10000:
    discount = total_bill * 10 / 100
else:
    discount = 0

final_bill = total_bill - discount

print("Total Bill =", total_bill)
print("Discount =", discount)
print("Final Bill =", final_bill)
```

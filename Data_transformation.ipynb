```python
# Numerical columns

num_cols = df.select_dtypes(include=["int64","float64"])

print(num_cols.columns)
```

```python
# Standard Scaling

from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()

standard_scaled = standard_scaler.fit_transform(num_cols)

print(standard_scaled[:5])
```

Standard Scaling is useful when data contains different scales and when algorithms assume normally distributed data.

```python
# Min Max Scaling

from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()

minmax_scaled = minmax_scaler.fit_transform(num_cols)

print(minmax_scaled[:5])
```

Min-Max Scaling is useful when features need to be transformed into a fixed range between 0 and 1.

```python
#2. Encoding techniques
# One Hot Encoding

onehot_cols = ["sex","race","income"]

onehot_df = pd.get_dummies(df,
                           columns=onehot_cols)

print(onehot_df.head())
```

```python
# Label Encoding

from sklearn.preprocessing import LabelEncoder

label_cols = [
    "workclass",
    "education",
    "marital_status",
    "occupation",
    "relationship",
    "native_country"
]

le = LabelEncoder()

for col in label_cols:

    df[col] = le.fit_transform(df[col])

print(df.head())
```

One-Hot Encoding

Pros:
Does not create an order among categories.

Works well for nominal categorical data.



Cons:
Increases the number of columns.

Label Encoding

Pros:
Simple and easy to implement.
Uses less memory.

Cons:
May introduce an artificial order among categories.


```python
# Creating new feature

df["age_hours"] = df["age"] * df["hours_per_week"]

print(df[["age", "hours_per_week", "age_hours"]].head())
```

```python
# Creating another feature

df["capital_total"] = df["capital_gain"] + df["capital_loss"]

print(df[["capital_gain",
          "capital_loss",
          "capital_total"]].head())
```

Feature 1: age_hours

This feature combines age and hours worked per week.
It may help capture the relationship between age and work effort.

Feature 2: capital_total

This feature combines capital gain and capital loss.
It provides overall capital-related information in a single feature.

```python
# Applying log transformation

df["log_fnlwgt"] = np.log1p(df["fnlwgt"])

print(df[["fnlwgt", "log_fnlwgt"]].head())
```

Log transformation was applied to the fnlwgt feature.

The fnlwgt column is positively skewed.
Applying log transformation helps reduce skewness and makes the distribution more balanced.

```python

import pandas as pd

df = pd.read_csv("sales_data_with_discounts.csv")

num_cols = df.select_dtypes(include="number")

print("Mean")
print(num_cols.mean())

print("\nMedian")
print(num_cols.median())

print("\nMode")
print(num_cols.mode())

print("\nStandard Deviation")
print(num_cols.std())

#The mean shows the average value of each numerical column.

#The median represents the middle value of the data.

##The mode shows the most frequently occurring value.

#The standard deviation indicates how much the values vary from the average value.

##These statistics help in understanding the overall distribution and behavior of the data.
```

```python
import matplotlib.pyplot as plt

num_cols.hist(figsize=(10,8))

plt.show()
```

```python
for col in num_cols.columns:

    plt.figure(figsize=(4,3))

    plt.boxplot(df[col])

    plt.title(col)

    plt.show()
```

```python
cat_cols = df.select_dtypes(include="object")

print(cat_cols.columns)
```

```python
for col in cat_cols.columns:

    plt.figure(figsize=(6,4))

    df[col].value_counts().plot(kind="bar")

    plt.title(col)

    plt.show()
```

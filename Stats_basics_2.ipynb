```python
# Importing pandas library

import pandas as pd
```

```python
# Loading patient dataset

patients = pd.read_csv("Patient_Data.csv")

#dataset information

patients.info()
```

```python
# Selecting billing related columns

billing_cols = patients[["PatientID", "Department", "Doctor", "BillAmount"]]

print(billing_cols)
```

```python
# Removing unnecessary columns

patients = patients.drop(["ReceptionistID", "CheckInTime"], axis=1)

print(patients.head())
```

```python
# Finding total bill amount for each department

dept_bill = patients.groupby("Department")["BillAmount"].sum()

print(dept_bill)
```

```python
# Removing duplicate patient records

patients = patients.drop_duplicates(subset="PatientID")

print(patients)
```

```python
# Filling missing bill amount with mean value

patients["BillAmount"] = patients["BillAmount"].fillna(patients["BillAmount"].mean())

print(patients)
```

```python
# Loading billing dataset

billing = pd.read_csv("Billing_Data.csv")

# Merging both datasets

merged_data = pd.merge(patients, billing, on="PatientID")

print(merged_data)
```

```python
# Creating new patient records

new_patients = pd.DataFrame({
    "PatientID":[111,112],
    "Name":["Ravi","Sita"],
    "Department":["Cardiology","Neurology"],
    "Doctor":["Dr. Smith","Dr. John"],
    "BillAmount":[5500,6500]
})

# Adding new records

all_patients = pd.concat([patients, new_patients], ignore_index=True)

print(all_patients)
```

```python
# Creating new billing columns

new_billing = pd.DataFrame({
    "InsuranceCovered":[2000,1500,2500,3000,1000],
    "FinalAmount":[3000,3500,5000,3200,4000]
})

# Adding new columns

final_data = pd.concat([patients, new_billing], axis=1)

print(final_data)
```

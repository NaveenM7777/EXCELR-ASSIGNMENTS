```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Cardiotocographic.csv")

print(df.head())
```

```python
df.info()
```

```python

print(df.isnull().sum())
```

```python
# Filling missing values with mean

df = df.fillna(df.mean())

print(df.isnull().sum())
```

```python
# Displaying data types

print(df.dtypes)
```

```python
# Detecting outliers using boxplots

for col in df.columns:

    plt.figure(figsize=(4,3))

    plt.boxplot(df[col])

    plt.title(col)

    plt.show()
```

```python

print(df.describe())
```

```python
# Calculating mean

print(df.mean())
```

```python
# Calculating median

print(df.median())
```

```python
# Calculating standard deviation

print(df.std())
```

```python
# Calculating IQR

Q1 = df.quantile(0.25)

Q3 = df.quantile(0.75)

IQR = Q3 - Q1

print(IQR)


#Observation:

#Mean, median and standard deviation were calculated for all numerical variables.

#The IQR helped in understanding the spread of the data.

#Some variables showed higher variation compared to others.
```

```python
# Creating histograms

df.hist(figsize=(12,10))

plt.show()

#Observation:

#Histograms were used to understand the distribution of numerical variables.

#Some variables appear normally distributed while others show skewness.

#A few variables contain extreme values.
```

```python
# Creating boxplots

for col in df.columns:

    plt.figure(figsize=(4,3))

    sns.boxplot(x=df[col])

    plt.title(col)

    plt.show()


#Observation:

#Boxplots helped identify outliers in several variables.

#Some columns contain extreme values that are far from the median.

#The spread of data varies across different variables.
```

```python
# Creating correlation heatmap

plt.figure(figsize=(12,8))

sns.heatmap(df.corr(), annot=True, cmap="coolwarm")

plt.show()

#Observation:

#The heatmap shows relationships between variables.

#Some variables have positive correlations while others have weak correlations.

#Strong correlations may be useful for further analysis.
```

```python
# Creating pair plot

sns.pairplot(df)

plt.show()

#Observation:

#Pair plots help visualize relationships between variables.

#Some variables show visible patterns and associations.

#The plots provide a deeper understanding of the dataset.
```

Conclusion:

The dataset was cleaned by handling missing values and checking data types.

Statistical measures such as mean, median, standard deviation and IQR were calculated.

Histograms and boxplots helped understand data distribution and identify outliers.

Correlation heatmaps and pair plots were used to study relationships between variables.

Overall, the exploratory analysis provided useful insights into the cardiotocographic dataset and can support further analysis.

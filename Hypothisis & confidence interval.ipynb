```python
import numpy as np
from scipy import stats
```

```python
data = [1.13, 1.55, 1.43, 0.92, 1.25,
        1.36, 1.32, 0.85, 1.07, 1.48,       #Data given in Question
        1.20, 1.33, 1.18, 1.22, 1.29]

n = len(data)

sample_mean = np.mean(data)

sample_std = np.std(data, ddof=1)

print("Sample Mean =", sample_mean)

print("Sample sd=", sample_std)
```

```python
# Finding t critical value for 99% confidence interval

t_value = stats.t.ppf(0.995, df=n-1)

print("t value =", t_value)
```

```python
# Calculating margin of error

margin_error = t_value * (sample_std / np.sqrt(n))

print("Margin of Error =", margin_error)
```

```python
# Calculating confidence interval using sample standard deviation

lower_limit = sample_mean - margin_error

upper_limit = sample_mean + margin_error

print("99% Confidence Interval")

print(lower_limit, upper_limit)
```

# Explanation:
# Sample standard deviation is used because the population
# standard deviation is not known.
#
# Since the sample size is small (n = 15),
# t-distribution is used instead of z-distribution.
#
# The confidence interval gives the range in which
# the true population mean is likely to fall with
# 99% confidence.

```python
# Given population standard deviation

population_std = 0.2
```

```python
# Finding z critical value for 99% confidence interval

z_value = stats.norm.ppf(0.995)

print("z value =", z_value)
```

```python
# Calculating margin of error

margin_error = z_value * (population_std / np.sqrt(n))

print("Margin of Error =", margin_error)
```

```python
# Calculating confidence interval using population standard deviation

lower_limit = sample_mean - margin_error

upper_limit = sample_mean + margin_error

print("99% Confidence Interval")

print(lower_limit, upper_limit)
```
